## Merge Busses and Substations from PowerWorld

Now add GPS coordinates for busses. See the [bus docs](https://pypsa.readthedocs.io/en/latest/user-guide/components.html#bus)

```python
# Get the full data from the PowerWorld AUX data 
buses = pd.read_csv('hawaii40_auxbus.csv')
buses.head()
```

```python
subs = pd.read_csv('hawaii40_auxsub.csv')
subs.head()
```

```python
subs = subs[['Number', 'Latitude', 'Longitude']]
subs = subs.rename(columns={'Number': 'SubNumber'})
```

```python
# Drop the Lat/Long from buses and get it from the Substation instead.
buses = buses.drop('Latitude', axis=1)
buses = buses.drop('Longitude', axis=1)
df = pd.merge(buses, subs, on='SubNumber')
df['Number'] = df['Number'].map(str) 
df = df.set_index('Number')
```

```python
df.head()
```

# Add GPS to pypsa data

```python
df.loc['1', 'Latitude'] 
```

```python
# Example of one data element
network.buses.loc['1', 'x']
```

```python
network.buses['x'] = df['Longitude']
network.buses['y'] = df['Latitude']
```

```python
# Add the name as a new field - this will stay with the buses dataframe during export/import
network.buses['BusName'] = df['Name'] # extra column not in spec
```

```python
fig, ax = plt.subplots(
    1, 1, subplot_kw={"projection": ccrs.EqualEarth()}, figsize=(8, 8)
)
network.plot.map(bus_sizes=5e-5, ax=ax, title="Hawaii 40 (with bus coordinates)");
```
